# Time series classification


```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt

import tensorflow as tf

from keras import backend as K
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM, CuDNNLSTM
from keras.layers import Dropout
from keras.layers import Activation

from keras.optimizers import Adam, SGD
from keras.models import load_model
from keras.callbacks import ModelCheckpoint
```

### Importamos el dataset

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

series = train.iloc[:,:-1]
targets = train.iloc[:,-1]
```

### Dividimos el dataset para training (66%) y validation (33%)

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val= train_test_split(series, targets, test_size=0.33, random_state= 42)
X_train = reshape(X_train)
X_val = reshape(X_val)
y_train = y_train.values
y_val = y_val.values
```

#### Algunas funciones

```python

def reshape(series): # reordena el array para que entre en la network con dimension (8040,1,260)
    series_np = series.values
    return np.reshape(series_np, (series_np.shape[0], 1,  series_np.shape[1]))


def auc(y_true, y_pred): # metrica que mide el "area under the curve"
    auc = tf.metrics.auc(y_true, y_pred)[1]
    K.get_session().run(tf.local_variables_initializer())
    return auc
```

## Modelos

He intentado resolver el problema de dos maneras diferentes:

#### Red LSTM
Para la primera solución he utilizado una red LSTM, ya que suelen dar buen resultado para series temporales de datos gracias a su arquitectura interna. Por lo general se usan para hacer predicciones sobre los siguientes valores en la serie, pero en este caso la tarea ha consistido en clasificar series enteras como reales o sintéticas, por ello he añadido al final una "Dense layer" encargada de la clasificación final teniendo en cuenta a cada serie al completo.

Teniendo en cuenta que el resultado se evalua en función del AUC, he añadido esta métrica como parámetro para seleccionar el 
modelo que maximizase esto durante el entrenamiento. Tras 400 epochs, el modelo se estabiliza con un AUC de 0.85 .
Sin embargo el valor de accuracy queda en 51%, lo cual me extraña mucho, por ello decidí buscar otra solución

#### Residual Networks
Después de investigar en internet he encontrado un paper (["Deep learning for time series classification: a review"](https://www.researchgate.net/publication/331474581_Deep_learning_for_time_series_classification_a_review)) en el que se comparan varios algoritmos para la clasificación de series temporales. En ese paper se ejecutan 9 algoritmos sobre 98 datasets distintos, y el más eficiente es Resnet, una red residual con varias convoluciones. Resnet usa categorical_crossentropy como función de loss, y devuelve un vector con las probabilidades para cada categoría. En nuestro caso, en el que solo hay dos categorías, la función loss se convierte en un "binary_crossentropy" 
El código está publicado en [github](https://github.com/hfawaz/dl-4-tsc), y lo que lo he utilizado para resolver el problema, obteniendo un mejor resultado (accuracy es del 87%) que con la red LSTM

### Red LSTM

```python
model = Sequential()
model.add(CuDNNLSTM(200, input_shape=(1, 260))) #LSTM layer diseñada para optimización con GPU
model.add(Dropout(0.4))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
# Two callbacks are created to monitor the auc and the accuracy
chk = ModelCheckpoint('best_model.pkl', monitor='val_auc', save_best_only=True, mode='max', verbose=1)
chk1 = ModelCheckpoint('best_model.pkl', monitor='val_acc', save_best_only=True, mode='max', verbose=1)

# Adam optimizer with custom learning rate
adam = Adam(lr=0.001)
```

```python
model.compile(loss='binary_crossentropy', optimizer=adam , metrics=[auc, "accuracy"])
```

```python
model.fit(X_train, y_train, validation_data=(X_val,y_val), epochs=200, callbacks=[chk,chk1])
```

#### Resultados

```python
ev = model.evaluate(X_train,y_train)
ev_val = model.evaluate(X_val,y_val)
print("loss: ", ev[0], " val_loss: ", ev_val[0])
print("auc: ", ev[1], " val_auc: ", ev_val[1])
print("acc: ", ev[2], " val_acc: ", ev_val[2])
```
